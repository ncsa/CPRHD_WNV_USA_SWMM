## Purpose: 

To highlight key SWMM Variables to extract from the simulation files on-the-fly to save disk space.

### Imports

```python
import pandas as pd
from swmmtoolbox import swmmtoolbox
import glob
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore", category=FutureWarning) # Since SWMMToolbox uses a deprecated function, supress the warning for it
```

### List All Variables

We will list the set of available extraction variables using swmmtoolbox's catalog feature. We'll do this for each individual scenario because they can potentially have different variables.
#### No Green Infrastructure

```python
ng_file = glob.glob('./simulation_files/*ng.out')[0]
swmmtoolbox.catalog(ng_file)
```

#### Rain Barrel

```python
rb_file = glob.glob('./simulation_files/*rb.out')[0]
swmmtoolbox.catalog(rb_file)
```

#### Rain Garden

```python
rg_file = glob.glob('./simulation_files/*rg.out')[0]
swmmtoolbox.catalog(rg_file)
```

### Check System Variables

We will extract and plot *roughly* the first year of the 'system' variables for each file.

```python
date = [i for i in range(298031//34)]

sim_types = {0:'No Green Infrastructure', 1: 'Rain Barrel', 2: 'Rain Garden'}

for var in [x for x in swmmtoolbox.catalog(rg_file) if 'system' in x]:
    print(var)
    fig, ax = plt.subplots(1,3, figsize=(28,8))
    fig.suptitle(var[2])
    count = 0
    for sim_type_file in [ng_file, rb_file, rg_file]:
        data = swmmtoolbox.extract(sim_type_file, var)
        data = data.iloc[:,0][:len(data)//34]
        ax[count].bar(date, data, width=1.0)
        ax[count].set_title(sim_types[count])
        count += 1
    plt.show()
    
```

### Plotting 'Node' Variables for Rain Barrel Scenario

We will plot *roughly* the first 5 years of the rain barrel scenario's node variables.

The node represents the rain barrel itself.

```python
node_variables = [x for x in swmmtoolbox.catalog(rb_file) if 'node' in x]
for var in node_variables:
    print(var)
    data = swmmtoolbox.extract(rb_file, var)
    data = data.iloc[0:32000]
    data.plot()
    plt.show()
```

### Compare 'Subcatch1' and 'System' data (No Green Infrastructure and Rain Garden)

In the models where there is only one subcatchment, ideally the Subcatch1 and System data will be equal. 

We will see if this is the case.

<br>

They don't have the exact same variables. 

For now we will compare the **'Rainfall'** variables. 

We will also compare Subcatch1's **'Evaporation_loss' + 'Infiltration_loss'** data to the system  **'Evaporation_infiltration'** data.

#### Compare Rainfall

```python
_sim_types = {'ng':'No Green Infrastructure', 'rg':'Rain Garden'}

for sim_type in [ng_file, rg_file]:
    name = _sim_types[(sim_type[sim_type.rfind('_')+1:sim_type.rfind('.')])]
    print(name)
    
    subcatch = swmmtoolbox.extract(sim_type, 'subcatchment,Subcatch1,Rainfall').iloc[:,0]
    system = swmmtoolbox.extract(sim_type, 'system,Rainfall,Rainfall').iloc[:,0]
    diff = subcatch - system
    
    print('Mean Difference:', diff.mean())
    print('Max Difference:', diff.max())
    print('Min Difference:', diff.min())
    print()
    
    plt.title(name)
    diff.plot()
    plt.ylabel('Difference (inches)')
    plt.show()
```

#### Compare 'Evaporation_loss' + 'Infiltration_loss' to 'Evaporation_infiltration'

```python
_sim_types = {'ng':'No Green Infrastructure', 'rg':'Rain Garden'}

for sim_type in [ng_file, rg_file]:
    name = (_sim_types[(sim_type[sim_type.rfind('_')+1:sim_type.rfind('.')])])
    print(name)
    
    subcatch_evap = swmmtoolbox.extract(sim_type, 'subcatchment,Subcatch1,Evaporation_loss').iloc[:,0]
    subcatch_infil = swmmtoolbox.extract(sim_type, 'subcatchment,Subcatch1,Infiltration_loss').iloc[:,0]
    
    system = swmmtoolbox.extract(sim_type, 'system,Evaporation_infiltration,Evaporation_infiltration').iloc[:,0]
    
    diff = subcatch_evap + subcatch_infil - system
    
    print('Mean Difference:', diff.mean())
    print('Max Difference:', diff.max())
    print('Min Difference:', diff.min())
    print()
    
    plt.title(name)
    diff.plot()
    plt.ylabel('Difference (inches)')
    plt.show()
```

### Compare 'Subcatch1 + Subcatch4' and 'System' data (Rain Barrel)

In the rain barrel scenario we add another subcatchment to act as the barrel. This will be added and the calculations above will be repeated.

#### Rainfall



```python
print('Rain Barrel')
sc1 = swmmtoolbox.extract(rb_file, 'subcatchment,Subcatch1,Rainfall').iloc[:,0]
sc4 = swmmtoolbox.extract(rb_file, 'subcatchment,Subcatch4,Rainfall').iloc[:,0]
system = swmmtoolbox.extract(rb_file, 'system,Rainfall,Rainfall').iloc[:,0]

diff = sc1 + sc4 - system
print('Mean Difference:', diff.mean())
print('Max Difference:', diff.max())
print('Min Difference:', diff.min())
print()

plt.title('Rain Barrel')
diff.plot()
plt.ylabel('Difference (inches)')
plt.show()
```

#### Evaporation_loss, Infiltration_loss, and Evaporation_infiltration

```python
sc1_evap = swmmtoolbox.extract(rb_file, 'subcatchment,Subcatch1,Evaporation_loss').iloc[:,0]
sc1_infil = swmmtoolbox.extract(rb_file, 'subcatchment,Subcatch1,Infiltration_loss').iloc[:,0]

sc4_evap = swmmtoolbox.extract(rb_file, 'subcatchment,Subcatch4,Evaporation_loss').iloc[:,0]
sc4_infil = swmmtoolbox.extract(rb_file, 'subcatchment,Subcatch4,Infiltration_loss').iloc[:,0]

system = swmmtoolbox.extract(rb_file, 'system,Evaporation_infiltration,Evaporation_infiltration').iloc[:,0]

diff = sc1_evap + sc1_infil + sc4_evap + sc4_infil - system

print('Mean Difference:', diff.mean())
print('Max Difference:', diff.max())
print('Min Difference:', diff.min())
print()

plt.title('Rain Barrel')
diff.plot()
plt.ylabel('Difference (inches)')
plt.show()
```

### Plot Analysis Conclusion

Although the mean difference is relatively low, this is because for the majority of the time, there is no rainfall. 

By looking at the max and min difference, we see that the differences are relatively large for hourly-level data.

Looking at the plots of the difference, we see that the peaks occur during periods of rainfall (which is reasonable as it's the only time a difference can really occur)

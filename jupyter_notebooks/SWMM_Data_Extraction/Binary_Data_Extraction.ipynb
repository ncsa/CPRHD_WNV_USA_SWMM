## Purpose: 

To highlight key SWMM Variables to extract from the simulation files on-the-fly to save disk space.

### Imports

```python
import pandas as pd
from swmmtoolbox import swmmtoolbox
import glob
import matplotlib.pyplot as plt
```

### List All Variables

We will list the set of available extraction variables using swmmtoolbox's catalog feature. We'll do this for each individual scenario because they can potentially have different variables.
#### No Green Infrastructure

```python
ng_file = glob.glob('./simulation_files/*ng.out')[0]
swmmtoolbox.catalog(ng_file)
```

#### Rain Barrel

```python
rb_file = glob.glob('./simulation_files/*rb.out')[0]
swmmtoolbox.catalog(rb_file)
```

#### Rain Garden

```python
rg_file = glob.glob('./simulation_files/*rg.out')[0]
swmmtoolbox.catalog(rg_file)
```

### Check System Variables

We will extract and plot the 'system' variables for each file.

```python
date = [i for i in range(12418)]

for var in [x for x in swmmtoolbox.catalog(rg_file) if 'system' in x]:
    fig, ax = plt.subplots(1,3, figsize=(28,8))
    fig.suptitle(var[2])
    count = 0
    for sim_type_file in [ng_file, rb_file, rg_file]:
        data = swmmtoolbox.extract(sim_type_file, var)
        data = data.iloc[:,0]
        print(data)
        ax[count].bar(date, data, width=1.0)
        count += 1
    plt.show()
    
```

### Purpose:

To plot the evaporation data for a few block groups over the course of 34 years.

```python
import matplotlib.pyplot as plt
import pandas as pd
import glob
```

```python
evap = pd.read_pickle('./evaporation_converted.pkl')
avg_evap = pd.read_pickle('../average_monthly_data/evaporation_converted.pkl')
```

```python
block_groups = {'260810146022': 'Kent, Michigan', '060375508003': 'Los Angeles, California', '170319800001': 'Chicago, Illinois', 
                '080310034021': 'Denver, Colorado', '482019801001': 'Harris, Texas', '120110502083': 'Broward, Florida', '360470220002': 'Kings, New York'
               }
```

#### Plotting

We will plot evaporation data for 7 block groups across the United States.

```python
for bg in ['260810146022', '060375508003', '170319800001', '080310034021', '482019801001', '120110502083', '360470220002']:
    fig, ax = plt.subplots(1,2,figsize=(16,8), sharey=True)
    fig.suptitle(block_groups[bg])
    ax[0].plot(evap.loc[bg])
    ax[0].set_title('Monthly Evaporation')
    print(evap.loc[bg])
    
    data = avg_evap.loc[bg]
    data = pd.concat([data] * 34)
    data.index = pd.DatetimeIndex(pd.date_range(start='1981-01-01', end='2014-12-01', freq='MS'))
    
    ax[1].plot(data)
    ax[1].set_title('Average Evaporation')
    plt.show()
    
```
